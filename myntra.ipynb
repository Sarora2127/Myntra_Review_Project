```python
import os  # helps in interacting with os
import time  # to introduce sleep while scraping
import requests  # HTTPS Requests
from bs4 import BeautifulSoup as bs # Pass HTML Content
from selenium import webdriver # Browser interact with dynamic content
from selenium.webdriver.common.by import By # How selenium should access content
```

```python
from selenium import webdriver
# Define the path to chrome webdriver executable
product = 'Souled Store Tshirt' 

# Initialize the chrome web driver
driver = webdriver.Chrome()

# Navigate to the URL
driver.get("https://www.myntra.com/kurta?rawQuery="+product)

# Once the page is loaded you can extract the page text
myntra_text = driver.page_source
myntra_html=bs(myntra_text,'html.parser')
```

```python
pclass = myntra_html.find_all("ul",{"class":"results-base"})
# find_all() is a method provided by BeautifulSoup to find all matching HTML elements based on the tag name and attributes you specify.
# syntax :- soup.find_all(tag_name, {attribute_name: value})
# Will return a list with one element — the first <ul> block only.
# If no match is found, it returns an empty list.
```

```python
for i in pclass:
    href = i.find_all('a',href=True)
    t = (href[0]['href'])
# Within the current <ul> (i), this finds all <a> tags that have an href attribute.
# Takes the first <a> tag from the list href[0]
# Accesses its href attribute

productLink = "https://www.myntra.com/"+t
productLink
```

```python
driver.get(productLink) # redirected to Specific Product Description
proRes = driver.page_source
```

```python
proRes_html = bs(proRes, "html.parser")
proRes_html
```

```python
import re

rating_count_div = proRes_html.find("div", {"class": "index-ratingsCount"})

if rating_count_div:
    rating_text = rating_count_div.get_text(strip=True)
    match = re.search(r"[\d.]+[kKmM]?", rating_text)  # Matches '3.4k', '500', etc.
    
    if match:
        raw_rating = match.group()

        # Convert to integer
        if 'k' in raw_rating.lower():
            rating_count = int(float(raw_rating.lower().replace('k', '')) * 1000)
        elif 'm' in raw_rating.lower():
            rating_count = int(float(raw_rating.lower().replace('m', '')) * 1000000)
        else:
            rating_count = int(raw_rating)

        print(f"Rating Count: {rating_count}")
```

```python
descriptors_container = proRes_html.find('div', class_='pdp-productDescriptorsContainer')
print(descriptors_container, end="\n")
product_details = {}
if descriptors_container:
    titles = descriptors_container.find_all('h4', class_='pdp-product-description-title')
    contents = descriptors_container.find_all('p', class_='pdp-product-description-content')

    for title, content in zip(titles, contents):
        section_name = title.get_text(strip=True)
        ul = content.find('ul')
        if ul:
            items = [li.get_text(strip=True) for li in ul.find_all('li')]
        else:
            items = content.get_text(separator="\n").strip().split('\n')
        product_details[section_name] = items
specs_container = proRes_html.find('div', class_='index-tableContainer')
if specs_container:
    rows = specs_container.find_all('div', class_='index-row')
    for row in rows:
        key_div = row.find('div', class_='index-rowKey')
        value_div = row.find('div', class_='index-rowValue')
        if key_div and value_div:
            key = key_div.get_text(strip=True)
            value = value_div.get_text(strip=True)
            product_details[key] = value
import pprint
pprint.pprint(product_details)
```

```python
review_html = bs(review_page,"html.parser")
```

```python
review = review_html.find_all("div",{"class":"detailed-reviews-userReviewsContainer"})
print(review)
```

```python
for i in review :
    user_rating = i.find_all("div",{"class":"user-review-main user-review-showRating"}) # Individual Customer Rating
    user_comment = i.find_all("div",{"class":"user-review-reviewTextWrapper"}) # individual comment/ review
    user_name = i.find_all("div",{"class":"user-review-left"})
```

```python
user_name
```

```python
for i in range(len(user_rating)): # number of ratings in user_rating list
    rating = user_rating[i].find("span",{"class":"user-review-starRating"}).get_text().strip()
    comment = user_comment[i].text
    name = user_name[i].find_all("span")[0].text
    date = user_name[i].find_all("span")[1].text
    print(f"Date:{date},Rating:{rating},Name:{name},comment:{comment}")
```
